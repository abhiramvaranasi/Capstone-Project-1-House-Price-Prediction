# Hypothesis Testing


```python
# Test Null Average Saleprice of SaleCondition is equal

housedf_EDA.SalePrice.groupby(housedf_EDA.SaleCondition).mean()
```

```python
from scipy.stats import f_oneway
```

```python
Abnorml=housedf_EDA[housedf_EDA.SaleCondition=='Abnorml']
AdjLand=housedf_EDA[housedf_EDA.SaleCondition=='AdjLand']
Alloca=housedf_EDA[housedf_EDA.SaleCondition=='Alloca']
Family=housedf_EDA[housedf_EDA.SaleCondition=='Family']
Normal=housedf_EDA[housedf_EDA.SaleCondition=='Normal']
Partial=housedf_EDA[housedf_EDA.SaleCondition=='Partial']
```

```python
f_oneway(Abnorml.SalePrice,AdjLand.SalePrice,Alloca.SalePrice,Family.SalePrice,Normal.SalePrice,Partial.SalePrice)
```

```python
# since p value =7.988268404991176e-44 is less than 0.05 so reject null
```

```python
# Test Null No Association between Sale Condition and GarageQual
```

```python
from scipy.stats import chi2_contingency
```

```python
pd.crosstab(housedf_EDA.SaleCondition,housedf_EDA.GarageQual)
```

```python
chi2_contingency(pd.crosstab(housedf_EDA.SaleCondition,housedf_EDA.GarageQual))
```

```python
# since pvalue is 0.45033351274609823 is greater than 0.05 so Failed to reject null
```

```python
# Test Null No Association between Sale condition and RoofStyle
```

```python
pd.crosstab(housedf_EDA.SaleCondition,housedf_EDA.RoofStyle)
```

```python
chi2_contingency(pd.crosstab(housedf_EDA.SaleCondition,housedf_EDA.RoofStyle))
```

```python
# since pvalue is 6.468740407630049e-08 is less than 0.05 so reject null
```

```python
X=housetrain_df.drop('SalePrice',axis=1)
y=housetrain_df.SalePrice
```

```python
housetest_df=housetest_df.drop(['level_0','index','Id'],axis=1)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression

from sklearn.model_selection import cross_val_score
```

```python
reg=LinearRegression()
```

```python
regmodel=reg.fit(X,y)
```

```python
regmodel.score(X,y)
```

```python
regpredict=regmodel.predict(X)
```

```python
regresid=y-regpredict
```

```python
np.sqrt(np.mean(regresid**2)) # RMSE
```
